## 문제 1
+ 다음 6 개의 Expression에 대해 Evaluation 결과 값을 출력하고, 해당 결과가 나온 이유에 대해 설명하시오
    + 1 and 2 and 3 and 4
    + 1 or 2 or 3 or 4
    + 1 and 2 or 3 and 4
    + (1 and 2) or (3 and 4)
    + 1 or 2 and 3 or 4
    + (1 or 2) and (3 or 4)

```python
print eval(repr(1 and 2 and 3 and 4))
print eval(repr(1 or 2 or 3 or 4))
print eval(repr(1 and 2 or 3 and 4))
print eval(repr((1 and 2) or (3 and 4)))
print eval(repr(1 or 2 and 3 or 4))
print eval(repr((1 or 2) and (3 or 4)))
```

> 이유 : repr()에 의해서 반환된 문자열은 eval()를 통해서 연산하고 값으로 바꿀 수 있다. repr()는 객체의 공식적인 표현으로 사용자가 읽기 편하게 표현해주는 str()과 달리 최대한 실제 객체처럼 표현하기 위해 사용된다.and는 true인 경우 그 다음 값도 확인해야 되고 or의 경우 false인 경우에 그 다음 값을 확인해야 한다. 숫자는 0이 false이며 그 외의 숫자는 true 값을 갖게 된다. 그러므로 위와 같은 결과값을 얻을 수 있다.

## 문제 2
+ 경로에 해당하는 문자열 1개를 입력 받아 그 안에 디렉토리 경로명과 파일명을 분리하여 리스트로 반환하는 함수 div_path(s)를 작성하시오.

    + 각 디렉토리와 파일을 구분하는 문자는 '/'로 가정한다.
    + 반환하는 리스트의 첫번째 원소는 디렉토리이고 두번째 원소는 파일명이다.

```python
def div_path(str) :
    s = 0
    e = 0
    filepath=""
    fileinfo=[]
    
    while(e != -1) :
        e = str.find('/')
        filepath = filepath+str[s:e+1]
        str = str[e+1:]
    filename = str
    fileinfo.append(filepath)
    fileinfo.append(filename)
    
    print fileinfo
    
div_path('/home/chulsoo/test.txt')
div_path('/usr/local/bin/python')
```

## 문제 3
+ 두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '합집합'을 반환하는 함수 list_union(lista, listb)를 작성하시오.
    + 인자로 전달하는 리스트 2 개에는 정수값만 들어간다고 가정하자.
    + 함수 내에서 새로운 리스트를 만들어 그 리스트 내에 인자로 받은 두 리스트의 모든 원소를 넣어 반환한다.
    + 반환하는 리스트에는 절대로 중복된 원소가 들어 있으면 안된다 (집합의 조건).
    + 반환하는 리스트는 정렬이 되어 있어야 한다.
    + 다음과 같은 실행 및 출력 결과가 도출되어야 한다.
        + list_union([1, 2, 3], [1, 2, 4])
            + [1, 2, 3, 4]
        + list_union([-10, -5, 0, -1], [100, 9, 0, 9])
            + [-10, -5, -1, 0, 9, 100]
        + list_union([0, 1, 2], [0, 1, 2])
            + [0, 1, 2]
    + [참고] 리스트(l) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 l.append(10) 이다.
    + [참고] 임의의 정수값 (x)이 리스트 (l) 내에 존재하는지 판단하는 방법은 x in l 이다.

```python
def list_union(lista, listb) :
    resultlist = []
    
    for i in range(0, len(lista)) :
        if lista[i] not in resultlist : 
            resultlist.append(lista[i])
    for j in range(0, len(listb)) :
        if listb[j] not in resultlist :
            resultlist.append(listb[j])        
    
    resultlist = sorted(resultlist)
    print resultlist

list_union([1, 2, 3], [1, 2, 4])
list_union([-10, -5, 0, -1], [100, 9, 0, 9])
list_union([0, 1, 2], [0, 1, 2])
```

## 문제 4
+ [프로젝트 오일러 문제 4] 앞에서부터 읽을 때나 뒤에서부터 읽을 때나 모양이 같은 수를 대칭수(palindrome)라고 부릅니다. 두 자리 수를 곱해 만들 수 있는 대칭수 중 가장 큰 수는 9009 (= 91 × 99) 입니다. 세 자리 수를 곱해 만들 수 있는 가장 큰 대칭수는 얼마입니까?

```python
from operator import eq
result = 0

for i in range(100, 999) :
    for j in range(i, 999) :
        num = i * j
        if eq(str(num)[::-1], str(num)) :
            if result < num :
                result = num

print result
```

## 문제 5
+ [프로젝트 오일러 문제 5 ] 1 ~ 10 사이의 어떤 수로도 나누어 떨어지는 가장 작은 수는 2520입니다. 그러면 1 ~ 20 사이의 어떤 수로도 나누어 떨어지는 가장 작은 수는 얼마입니까?

```python
def gcd(a, b):
    while (b != 0):
        temp = a % b
        a = b
        b = temp
    return a

lcm = 1
for x in range(2, 20) :
    lcm = x * lcm / gcd(x, lcm)

print lcm
```

## 문제 6
+ [프로젝트 오일러 문제 6] 1부터 10까지 자연수를 각각 제곱해 더하면 다음과 같습니다 (제곱의 합).
        12 + 22 + ... + 102 = 385
1부터 10을 먼저 더한 다음에 그 결과를 제곱하면 다음과 같습니다 (합의 제곱).
        (1 + 2 + ... + 10)2 = 552 = 3025
따라서 1부터 10까지 자연수에 대해 "합의 제곱"과 "제곱의 합" 의 차이는 3025 - 385 = 2640 이 됩니다.
그러면 1부터 100까지 자연수에 대해 "합의 제곱"과 "제곱의 합"의 차이는 얼마입니까?

```python
pos = 0
sop = 0

for i in range(1, 101) :
    pos = pos + i*i
    sop = sop + i
sop = sop*sop

print sop-pos
```

## 문제 7
+ [Incremental Project] 웹 URL로 지정된 웹페이지를 문자열로 가져와 1) 모든 HTML 태그를 제외한 순수 텍스트 문자열만을 걸러내고, 2) 그 순수 텍스트 문자열 안에 존재하는 단어가 총 몇개인지 출력하는 프로그램을 작성하시오.
    + HTML 태그는 다음 조건을 가지는 것이다.
        + < 로 시작하여 >로 끝난다.
    + 단어를 나누는 기준은 오로지 공백문자 (whitespace)이다.
        + 즉, 위 순수 텍스트 문자열인 "로그인 유지"에서 분리한 단어는 "로그인"과 "유지"이다.
    + 다음은 URL로 지정된 웹페이지를 특정 문자열로 가져오는 코드이다.
        + 즉, 숙제 코드는 아래 코드로 부터 시작해야 한다.

```python
import urllib2
source = urllib2.urlopen("http://cse.kut.ac.kr/").read()

while(1) :
    if source.find('<!-') == -1 :
        break;
    source = source.replace(source[source.find('<!-'):source.find('->')+2:],'')

while(1) :
    if source.find('<') == -1 :
        break;
    source = source.replace(source[source.find('<'):source.find('>')+1:],'')
      
wordlist = source.split()

print "단어 수 : " + str(len(wordlist)) + "개"

for i in range(0, len(wordlist)) : 
    print wordlist[i]
```
